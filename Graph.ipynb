```python
# 1. Representasi Graf dengan Adjacency List

# Membuat graf tak berarah menggunakan adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Menampilkan graf
def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

print_graph(graph)
```

```python
# 2. Traversal Graf menggunakan BFS (Breadth-First Search)

from collections import deque

def bfs(graph, start):
    visited = set()
    queue = deque([start])

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            print(vertex, end=" ")
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited]) # Menambahkan tetangga yang belum dikunjungi ke antrian

# Contoh pemanggilan
print("Urutan BFS:")
bfs(graph, 'A')
```

```python
# 3. Traversal Graf menggunakan DFS (Depth-First Search)

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()

    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs(graph, neighbor, visited)

# Contoh pemanggilan
dfs(graph, 'A')
```

```python
# Fungsi untuk menghitung jumlah tetangga dari tiap simpul dalam graf
def count_neighbors(graph):
    result = {}
    for node in graph:
        # Menghitung panjang daftar tetangga untuk node tersebut
        result[node] = len(graph[node])
    return result

# --- Contoh Penggunaan ---
# Menggunakan contoh graf dari modul
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

hasil = count_neighbors(graph)
print(hasil)
# Output: {'A': 2, 'B': 3, 'C': 2, 'D': 1, 'E': 2, 'F': 2}
```

```python
# LATIHAN SOAL

from collections import deque

# Soal 1: Menampilkan graf
def print_graph(graf):
    for node in graf:
        print(f"{node} -> {graf[node]}")

# Soal 3: Modifikasi BFS agar mengembalikan list urutan kunjungan
def bfs_list(graph, start):
    visited = set()
    queue = deque([start])
    order = []  # Menyimpan urutan kunjungan

    while queue:
        vertex = queue.popleft()
        if vertex not in visited:
            order.append(vertex)
            visited.add(vertex)
            queue.extend([neighbor for neighbor in graph[vertex] if neighbor not in visited])

    return order

# Soal 2: Implementasi DFS rekursif untuk mencetak urutan
def dfs_print(graph, start, visited=None):
    if visited is None:
        visited = set()
    if start not in visited:
        print(start, end=" ")
        visited.add(start)
        for neighbor in graph[start]:
            dfs_print(graph, neighbor, visited)

# Soal 4: Menemukan satu jalur dari start ke end menggunakan DFS
def find_path(graph, start, end, visited=None, path=None):
    if visited is None:
        visited = set()
    if path is None:
        path = []

    visited.add(start)
    path = path + [start]

    if start == end:
        return path

    for neighbor in graph[start]:
        if neighbor not in visited:
            new_path = find_path(graph, neighbor, end, visited, path)
            if new_path:  # Jika jalur ditemukan
                return new_path

    return None  # Jika tidak ada jalur

# Soal 5: Mengecek apakah graf saling terhubung (connected)
def is_connected(graph):
    if not graph:
        return True

    # Ambil simpul pertama sebagai titik mulai
    start_node = list(graph.keys())[0]

    # Lakukan penelusuran (DFS/BFS) untuk mencatat semua simpul yang bisa dijangkau
    visited = set()

    def dfs_helper(node):
        visited.add(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                dfs_helper(neighbor)

    dfs_helper(start_node)

    # Jika jumlah simpul yang dikunjungi sama dengan total simpul, berarti graf terhubung
    return len(visited) == len(graph)

# PENGUJIAN

# Soal 1: Membuat graf tak berarah baru (5 simpul: A, B, C, D, E & 7 sisi)
# Sisi: (A-B), (A-C), (B-C), (B-D), (C-D), (C-E), (D-E)
my_graph = {
    'A': ['B', 'C'],
    'B': ['A', 'C', 'D'],
    'C': ['A', 'B', 'D', 'E'],
    'D': ['B', 'C', 'E'],
    'E': ['C', 'D']
}

print("=== 1. Tampilan Graf Baru (5 Simpul, 7 Sisi) ===")
print_graph(my_graph)
print("\n" + "="*40 + "\n")

print("=== 2. Traversal dari Simpul A ===")
print("Eksplorasi DFS: ", end="")
dfs_print(my_graph, 'A')
print()

# Menggunakan fungsi BFS hasil modifikasi dari soal 3
urutan_bfs = bfs_list(my_graph, 'A')
print(f"Eksplorasi BFS (Hasil Soal 3): {urutan_bfs}")
print("\n" + "="*40 + "\n")

print("=== 4. Pencarian Jalur (find_path) ===")
jalur = find_path(my_graph, 'A', 'E')
print(f"Jalur dari A ke E: {jalur}")
print("\n" + "="*40 + "\n")

print("=== 5. Pengecekan Keterhubungan Graf ===")
print(f"Apakah graf saling terhubung? {is_connected(my_graph)}")

# Pembuktian dengan graf terputus (disconnected)
graf_terputus = {
    'A': ['B'],
    'B': ['A'],
    'C': ['D'],
    'D': ['C'],
    'E': []
}
print(f"Apakah graf_terputus saling terhubung? {is_connected(graf_terputus)}")
```
