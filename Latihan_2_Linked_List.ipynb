```python
# Soal Nomor 1

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None
        self.prevval = None


class DoubleLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        printval = self.headval
        while printval is not None:
            print(printval.dataval)
            printval = printval.nextval

    def AddInEnd(self, newdata):
        NewNode = Node(newdata)
        if self.headval is None:
            self.headval = NewNode
            return

        last = self.headval
        while last.nextval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.prevval = last

#Penyisipan di tengah
    def InsertAfter(self, key, newdata):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                break
            temp = temp.nextval

        if temp is None:
            print("Key tidak ditemukan")
            return

        NewNode = Node(newdata)
        NewNode.nextval = temp.nextval
        NewNode.prevval = temp

        if temp.nextval is not None:
            temp.nextval.prevval = NewNode

        temp.nextval = NewNode

#Hapus dengan Kata Kunci
    def RemoveKey(self, key):
        temp = self.headval

        while temp is not None:
            if temp.dataval == key:
                break
            temp = temp.nextval

        if temp is None:
            print("Data tidak ditemukan")
            return

        # jika node pertama
        if temp.prevval is None:
            self.headval = temp.nextval
            if self.headval is not None:
                self.headval.prevval = None
        else:
            temp.prevval.nextval = temp.nextval
            if temp.nextval is not None:
                temp.nextval.prevval = temp.prevval

        temp = None


Li = DoubleLinkedList()

Li.AddInEnd("Heeseung")
Li.AddInEnd("Jaemin")
Li.AddInEnd("Mingyu")

print("Sebelum penyisipan:")
Li.listprint()
print("========================================")

#Setelah disisipkan
Li.InsertAfter("Jaemin", "Jeno")

print("Setelah disisipkan:")
Li.listprint()
print("========================================")

#hapus berdasarkan kata kunci
Li.RemoveKey("Mingyu")

print("Setelah removekey:")
Li.listprint()
```

```python
# Soal Nomor 2

class Node:
    def __init__(self, dataval=None):
        self.dataval = dataval
        self.nextval = None


class CircularLinkedList:
    def __init__(self):
        self.headval = None

    def listprint(self):
        if self.headval is None:
            return

        printval = self.headval
        while True:
            print(printval.dataval)
            printval = printval.nextval
            if printval == self.headval:
                break

    def AddInEnd(self, newdata):
        NewNode = Node(newdata)

        if self.headval is None:
            self.headval = NewNode
            NewNode.nextval = self.headval
            return

        last = self.headval
        while last.nextval != self.headval:
            last = last.nextval

        last.nextval = NewNode
        NewNode.nextval = self.headval

#Hapus node di akhir
    def RemoveEnd(self):
        if self.headval is None:
            return

        if self.headval.nextval == self.headval:
            self.headval = None
            return

        prev = None
        last = self.headval

        while last.nextval != self.headval:
            prev = last
            last = last.nextval

        prev.nextval = self.headval
        last = None


Li = CircularLinkedList()

Li.AddInEnd("Heeseung")
Li.AddInEnd("Jaemin")
Li.AddInEnd("Mingyu")

print("Sebelum penyisipan:")
Li.listprint()
print("==============================")

Li.AddInEnd("Sunoo")

print("Setelah penyisipan di akhir:")
Li.listprint()
print("==============================")

Li.RemoveEnd()

print("Setelah hapus node terakhir:")
Li.listprint()
```
