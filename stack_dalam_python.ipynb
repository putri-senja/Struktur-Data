```python
# Stack dengan List

#python program to demonstrate stack implementation using list
stack = []

#append() function to push element in the stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack')
print(stack)

#pop() function to pop element from stack in LIFO order
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped:')
print(stack)

#uncommeting print(stack.pop()) will cause an IndexError as the stack is now empty
```

```python
#Stack dengan Collections.deque

#Python program to demonstrate stack implementation using collections.deque
from collections import deque

stack = deque()

#append() function to push element in the stack
stack.append('a')
stack.append('i')
stack.append('u')
stack.append('e')
stack.append('o')

print('Initial stack:')
print(stack)

#pop() function to pop element from stack in LIFO order
print('\nElements popped from stack:')
print(stack.pop())
print(stack.pop())
print(stack.pop())

print('\nStack after elements are popped:')
print(stack)

#uncommenting print(stack.pop()) will cause an IndexError as the stack is now empty
```

```python
#Stack dengan queue.LifoQueue

#python program to demonstrate stack implementation using queue module
from queue import LifoQueue

#Initializing a stack
stack = LifoQueue(maxsize=5)

#qsize() show the number of elements in the stack
print(stack.qsize())

#put() function to push element in the stack
stack.put('a')
stack.put('i')
stack.put('u')
stack.put('e')
stack.put('o')

print("Full: ", stack.full())
print("Size: ", stack.qsize())

#get() function to pop element from stack in LIFO order
print('\nElements popped from the stack')
print(stack.get())
print(stack.get())
print(stack.get())

print("\nEmpty: ", stack.empty())
```

```python
#Stack dengan Single Linked List

class Node:
  def __init__(self, value):
    self.value = value
    self.next = None

class Stack:

  #Initializing a stack, Use a dummy node, which is easier for handling edge cases.
  def __init__(self):
    self.head = Node("head")
    self.size = 0

  #String representation of the stack
  def __str__(self):
    cur = self.head.next
    out = ""
    while cur:
      out += str(cur.value) + "->"
      cur = cur.next
    return out[:-2]

  #Get the current size of the stack
  def getSize(self):
    return self.size

  #Check if the stack is empty
  def isEmpty(self):
    return self.size == 0

  #Get the top item of the stack
  def peek(self):
    #Sanitary check to see if we are peeking an empty stack
    if self.isEmpty():
      raise Exception("Peeking from an empty stack")
    return self.head.next.value

  #Push a value into the stack
  def push(self, value):
    node = Node(value)
    node.next = self.head.next
    self.head.next = node
    self.size += 1

  #Remove a value from the stack and return
  def pop(self):
    if self.isEmpty():
      raise Exception("Popping from an empty stack")
    remove = self.head.next
    self.head.next = self.head.next.next
    self.size -= 1
    return remove.value

#Driver code
if __name__ == "__main__":
  stack = Stack()
  for i in range(1, 11):
    stack.push(i)
  print(f"Stack: {stack}")

  for  _ in range(1, 6):
    remove = stack.pop()
    print(f"Pop: {remove}")
  print(f"Stack: {stack}")
```

### Penjelasan Kode Sel `uROAsTkxjuTy`: Stack dengan Single Linked List (Implementasi Kustom)

Sel kode ini menyediakan implementasi stack kustom dari awal menggunakan konsep *single linked list*. Ini adalah cara yang lebih mendalam untuk memahami bagaimana stack bekerja di tingkat fundamental.

#### Kelas `Node`
*   **`class Node:`**: Merepresentasikan sebuah elemen tunggal dalam linked list.
*   **`__init__(self, value)`**: Konstruktor `Node` yang menerima sebuah `value` dan menginisialisasi `self.next` menjadi `None`. `self.next` akan menunjuk ke node berikutnya dalam list.

#### Kelas `Stack`
*   **`class Stack:`**: Mengimplementasikan fungsionalitas stack.
*   **`__init__(self):`**: Konstruktor stack.
    *   **`self.head = Node("head")`**: Membuat 'dummy node' bernama "head". Ini adalah praktik umum dalam linked list untuk menyederhanakan penanganan kasus tepi (seperti stack kosong) karena `self.head` selalu ada dan tidak menyimpan data stack yang sebenarnya.
    *   **`self.size = 0`**: Menginisialisasi ukuran stack menjadi 0.
*   **`__str__(self):`**: Metode khusus untuk representasi string dari stack, memungkinkan Anda mencetak objek stack secara langsung untuk melihat isinya. Ini akan menampilkan elemen dari atas ke bawah, dipisahkan oleh `->`.
*   **`getSize(self):`**: Mengembalikan jumlah elemen dalam stack.
*   **`isEmpty(self):`**: Mengembalikan `True` jika stack kosong (`self.size == 0`), `False` jika tidak.
*   **`peek(self):`**: Mengembalikan nilai elemen paling atas dari stack tanpa menghapusnya. Jika stack kosong, akan muncul `Exception`.
*   **`push(self, value):`**: Menambahkan elemen baru ke bagian atas stack.
    *   `node = Node(value)`: Membuat node baru untuk nilai yang diberikan.
    *   `node.next = self.head.next`: Node baru menunjuk ke node yang sebelumnya ada di bagian atas stack (yaitu, node yang ditunjuk oleh `self.head.next`).
    *   `self.head.next = node`: 'Head' dummy node sekarang menunjuk ke node baru, menjadikannya elemen paling atas.
    *   `self.size += 1`: Menambah ukuran stack.
*   **`pop(self):`**: Menghapus dan mengembalikan elemen paling atas dari stack.
    *   `if self.isEmpty(): raise Exception(...)`: Memeriksa apakah stack kosong untuk mencegah kesalahan.
    *   `remove = self.head.next`: Menyimpan referensi ke node yang akan dihapus (node paling atas).
    *   `self.head.next = self.head.next.next`: 'Head' dummy node sekarang menunjuk ke node berikutnya setelah node yang akan dihapus, secara efektif menghapus node paling atas dari linked list.
    *   `self.size -= 1`: Mengurangi ukuran stack.
    *   `return remove.value`: Mengembalikan nilai dari node yang dihapus.
